```python
import pandas as pd
```

```python
df=pd.read_csv("marketing_funnel_sample.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df.columns=df.columns.str.strip()
```

```python
print(df.isnull().sum())
```

```python
df["category_code"]=df["category_code"].fillna(0)
df["brand"]=df["brand"].fillna(0)
```

```python
df=df.drop_duplicates()
```

```python
df["date"]=pd.to_datetime(df["date"])
df["event_time"]=pd.to_datetime(df["event_time"])
```

```python
df["hour"]=df["event_time"].dt.hour
df["weekday"]=df["event_time"].dt.day_name()
```

```python
df.info()
```

```python
df["price"]=pd.to_numeric(df["price"])
df=df[df["price"]>=0]
```

```python
df.to_csv("cleaned_marketing_funnel.csv",index=False)
```
